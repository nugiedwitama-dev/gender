```python
def extract_images(path,gender,i):
    img = cv2.imread(path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = haar.detectMultiScale(gray,1.5,5)
    for x,y,w,h in faces:
        roi = img[y:y+h,x:x+w]
        if gender == 'male': 
            cv2.imwrite('./data/crop/male_crop/{}_{}.png'.format(gender,i),roi)
        else: 
            cv2.imwrite('./data/crop/female_crop/{}_{}.png'.format(gender,i),roi)
```

```python
extract_images(femalepath[0], 'female',1)
```

```python
for i, path in enumerate(femalepath):
    try:
        extract_images(path,'female',i)
        print("{}/{} berhasil diproses".format(i,len(femalepath)))
    
    except:
        print("{}/{} gagal diproses".format(i,len(femalepath)))
```

## Membuat struktur wajah dengan open cv

```python
def resize_img(path_to_resize):
    try:
        
        img = cv2.imread(path_to_resize)
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        size = gray.shape[0]
        if size >= 100:
            gray_re = cv2.resize(gray,(100,100),cv2.INTER_AREA)
        else:
            gray_re = cv2.resize(gray,(100,100),cv2.INTER_CUBIC)
        return gray_re.flatten()
    except:
        return None
```

## Feature Extraction - PCA Eigenface

```python
data = np.load('./data/data_100_100_norm.npz')
```

```python
X = data['arr_0'] #variabel independen
y = data['arr_1'] #target value
```

```python
X.shape, y.shape
```

## Eigen Image

```python
X1 = X -X.mean(axis=0)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=None, whiten=True,svd_solver='auto')
```

```python
x_pca = pca.fit_transform(X1)
```

```python
x_pca.shape
```

```python
eigen_ratio = pca.explained_variance_ratio_
eigen_ratio_cum = np.cumsum(eigen_ratio)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(eigen_ratio[:200],'r>--')
plt.xlabel=('no of components')
plt.tlabel=('explained variance ratio')
plt.subplot(1,2,2)
plt.xlabel=('no of components')
plt.xlabel=('cumulative explained variance ratio')
plt.plot(eigen_ratio_cum[:200],'r>--')
plt.show()
```

conclusion : using elbow method, consider number of component is between 25-30
> since if i consider components between 25 to 30 the explained variance is arround 75% so, in order 80% variance i am considering 50 components

```python
pca_50 = PCA(n_components=50, whiten=True,svd_solver='auto')
x_pca_50 = pca_50.fit_transform(X1)
```

```python
x_pca_50.shape
```

```python
pickle.dump(pca_50,open('./model/pca_50.pickle','wb'))
```

```python
x_inv_trans  = pca_50.inverse_transform(x_pca_50)
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
x_train,x_test,y_train,y_test = train_test_split(X,y,train_size=0.2,stratify=y)
```

## Training Machine Learning Model

```python
from sklearn.svm import SVC
model = SVC(C=1.0,kernel='rbf',gamma=0.01, probability=True )
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

## Evaluasi Model

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn import metrics
y_pred = model.predict(x_test)
y_prob = model.predict_proba(x_test)
```
